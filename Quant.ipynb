## Loading and Working with CSV Files

```python
# Load CSV file
df = pd.read_csv('MSFT.csv')

# Convert Date column to datetime
df['Date'] = pd.to_datetime(df['Date'])

# View first few rows
df.head()
```

```python
df_filtered = df[(df['Date'] >= '2020-01-01') & (df['Date'] <= '2020-12-31')]
```

```python
df_filtered
```

```python
x = df_filtered.set_index('Date')['Close']
x.describe()
```

```python
x
```

```python
plt.plot(x.index, x.values)
plt.xlabel('Date')
plt.ylabel('Close Price')
plt.xticks(rotation=45)
plt.tight_layout()
```

```python
R = x.pct_change()[1:]
```

```python
plt.hist(R, bins=20)
plt.xlabel('Returns')
plt.ylabel('Frequency')
plt.title('Msft Returns')
plt.show()
```

```python
np.mean(R)
```

```python
np.std(R)
```
